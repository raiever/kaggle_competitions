```python
def find_prime(number):
    for i in reversed(range(int(np.sqrt(number))+1)):
        print('i: ', i)
        if i == 1:
            return True
            break
        elif i == 2 and number % i == 0:
            return False
            break
        elif number % i == 0 and i != 1:
            number = int(number / i)
            print('number: ', number)
            return find_prime(number)
        else:
            return True
            break
```
